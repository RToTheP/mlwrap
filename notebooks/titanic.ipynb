```python
import pandas as pd
from tests.datasets import TitanicDataset

titanic = TitanicDataset()
df = pd.concat([titanic.X, titanic.y], axis=1)
df
```

```python
df.info(verbose=True)
```

```python
df.describe(include='all')
```

How many missing values do we have?

```python
df.isna().mean()
```

There are a fair few missing values in the `cabin`, `boat`, `body`, and `home.dest` columns.

Let's clean up the data

```python
from mlwrap import runners

df_clean = runners.clean(df)

df_clean.isna().mean()
```

```python
df_clean.columns = df.columns
df_clean
```

Now let's train a model on the cleaned data

```python
from mlwrap import scores
from mlwrap.config import MLConfig
from mlwrap.enums import ProblemType

config = MLConfig(problem_type=ProblemType.Classification, model_feature_id='survived')

results = runners.train(config, df)
scores.print_scores(results.scores)
```
